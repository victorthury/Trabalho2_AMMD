```python
import numpy as np
import pandas as pd
np.random.seed(42)

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

for i in [0,9,13]:
    si = plt.style.available[i]
    plt.style.use(si)
```

```python
red_df = pd.read_csv('../data/wine_quality/winequality-red.csv', sep=';')
red_df.head()
```

```python
white_df = pd.read_csv('../data/wine_quality/winequality-white.csv', sep=';')
white_df.head()
```

```python
def to_name(i):
    l = [
        'one',
        'two',
        'three',
        'four',
        'five',
        'six',
        'seven',
        'eght',
        'nine', 
        'ten'
    ]
    return l[int(i)]
```

```python
df = red_df.copy()
df.quality = df.quality.apply(to_name)
```

```python
sns.pairplot(df, hue='quality');
```

## Data Pre processing

```python
def values2zscores(df):
    zdf = df.copy()
    for cname in df:
        mean, std = df[cname].mean(), df[cname].std()
        zdf[cname] = (df[cname] - mean)/std
        
    return zdf

def normalize(df):
    ndf = df.copy()
    for cname in df:
        maxv, minv = df[cname].max(), df[cname].min()
        ndf[cname] = (df[cname] - minv)/(maxv-minv)
        
    return ndf
```

```python
red_y = red_df.quality
red_x = red_df.drop(columns=['quality'])

white_y = white_df.quality
white_x = white_df.drop(columns=['quality'])
```

### Zsocore

```python
red_z = values2zscores(red_x)
red_z.head()
```

```python
white_z = values2zscores(white_x)
white_z.head()
```

## Visualize

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
rpc = pca.fit_transform(red_z)
```

```python
pca = PCA(n_components=2)
wpc = pca.fit_transform(white_z)
```

```python
plt.scatter(wpc.T[0], wpc.T[1])
plt.scatter(rpc.T[0], rpc.T[1])

plt.show()
```

```python
plt.scatter(wpc.T[0], white_y)
plt.scatter(rpc.T[0], red_y)

plt.show()
```

## Try some Classification Models

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
```

```python
rdf = RandomForestClassifier()
csvm_lin = svm.SVC(kernel='linear', C=1)
csvm_rbf = svm.SVC(kernel='rbf', C=1)
gnb = GaussianNB()

models = [
    ('Random Forest', rdf), 
    ('Linar SVM', csvm_lin),
    ('RBF SVM', csvm_rbf),
    ('GaussianNB', gnb)
]
```

```python
print('Zscore\n')

for mname, model in models:
    red_c = cross_val_score(model, red_z, red_y, cv=5)
    white_c = cross_val_score(model, white_z, white_y, cv=5)
    print(f'{mname}:')
    print(f'\t[tred whine]  {np.mean(red_c)}\n\t[white whine] {np.mean(white_c)}')
#     print(f'\t{red_c, white_c}\n')
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
model = svm.SVC(kernel='rbf', C=1)
model.fit(red_z, red_y)
red_p = model.predict(red_z)

print(classification_report(red_y, red_p))
```
